```python
import re

def extract_value_and_unit(text):
    if not isinstance(text, str):
        return None, None

    # Match all numeric values (e.g., 468, 337.5, .3)
    number_matches = re.findall(r'\d+(?:\.\d+)?|\.\d+', text)

    # Match the unit that follows the last number in the sentence
    unit_match = re.search(r'(?:\d+(?:\.\d+)?|\.\d+)\s*([a-zA-Zμ/().\-+⁻−0-9 ]+)', text)

    if number_matches and unit_match:
        first_number = float(number_matches[0])
        unit = unit_match.group(1).strip()
        return first_number, unit

    return None, None
```

```python
print(extract_value_and_unit("468, 337, 268 mAh g(-1)"))          # ➜ (468.0, 'mAh g(-1)')
print(extract_value_and_unit("up to 90 mAh/(g LCO)"))             # ➜ (90.0, 'mAh/(g LCO)')
print(extract_value_and_unit("around 135 mAh g(-1)"))             # ➜ (135.0, 'mAh g(-1)')
print(extract_value_and_unit("greater than 1000.5 mAh/g"))        # ➜ (1000.5, 'mAh/g')
```

```python
#apply function
filtered_Li4Ti5O12['Capacity_Value'], filtered_Li4Ti5O12['Capacity_Units'] = zip(*filtered_Li4Ti5O12['Capacity'].apply(extract_value_and_unit))
filtered_LiCoO2['Capacity_Value'], filtered_LiCoO2['Capacity_Units'] = zip(*filtered_LiCoO2['Capacity'].apply(extract_value_and_unit))
filtered_LiFePO4['Capacity_Value'], filtered_LiFePO4['Capacity_Units'] = zip(*filtered_LiFePO4['Capacity'].apply(extract_value_and_unit))
filtered_TiO2['Capacity_Value'], filtered_TiO2['Capacity_Units'] = zip(*filtered_TiO2['Capacity'].apply(extract_value_and_unit))
filtered_MoS2['Capacity_Value'], filtered_MoS2['Capacity_Units'] = zip(*filtered_MoS2['Capacity'].apply(extract_value_and_unit))


filtered_Li4Ti5O12['C-rate_Value'], filtered_Li4Ti5O12['C-rate_Units'] = zip(*filtered_Li4Ti5O12['C-rate'].apply(extract_value_and_unit))
filtered_LiCoO2['C-rate_Value'], filtered_LiCoO2['C-rate_Units'] = zip(*filtered_LiCoO2['C-rate'].apply(extract_value_and_unit))
filtered_LiFePO4['C-rate_Value'], filtered_LiFePO4['C-rate_Units'] = zip(*filtered_LiFePO4['C-rate'].apply(extract_value_and_unit))
filtered_TiO2['C-rate_Value'], filtered_TiO2['C-rate_Units'] = zip(*filtered_TiO2['C-rate'].apply(extract_value_and_unit))
filtered_MoS2['C-rate_Value'], filtered_MoS2['C-rate_Units'] = zip(*filtered_MoS2['C-rate'].apply(extract_value_and_unit))

filtered_Li4Ti5O12['Voltage_Value'], filtered_Li4Ti5O12['Voltage_Units'] = zip(*filtered_Li4Ti5O12['Voltage'].apply(extract_value_and_unit))
filtered_LiCoO2['Voltage_Value'], filtered_LiCoO2['Voltage_Units'] = zip(*filtered_LiCoO2['Voltage'].apply(extract_value_and_unit))
filtered_LiFePO4['Voltage_Value'], filtered_LiFePO4['Voltage_Units'] = zip(*filtered_LiFePO4['Voltage'].apply(extract_value_and_unit))
filtered_TiO2['Voltage_Value'], filtered_TiO2['Voltage_Units'] = zip(*filtered_TiO2['Voltage'].apply(extract_value_and_unit))
filtered_MoS2['Voltage_Value'], filtered_MoS2['Voltage_Units'] = zip(*filtered_MoS2['Voltage'].apply(extract_value_and_unit))
```
